In this attempt I am using a csv agent. It translates the query to pandas query format. It works pretty well but it costs a lot.

```python
queries = ["The youngest.","how many germans?", "Tell me the driver name with oldest birthdate.", "driver uses the number 14","german and less than 25 years old","uses the number 14 and less than 20 years"]
```

```python
from langchain.agents import create_csv_agent
from langchain.llms import OpenAI
from langchain.callbacks import get_openai_callback

with get_openai_callback() as cb:

    open_ai = OpenAI(temperature=0)

    print(open_ai.model_name)

    agent = create_csv_agent(open_ai, 'drivers_full.csv', verbose=True)

    for query in queries: 
        agent.run(query)

    print(cb)

    
```
